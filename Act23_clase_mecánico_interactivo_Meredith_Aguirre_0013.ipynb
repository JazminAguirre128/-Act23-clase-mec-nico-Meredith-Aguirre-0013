# Mecanico

```python
class Mecanico:
    def __init__(self, nombre, especialidad, herramienta_favorita):
        self.nombre = nombre
        self.especialidad = especialidad
        self.herramienta_favorita = herramienta_favorita
        self.taller_abierto = False
        self.autos_en_espera = 0

    def abrir_taller(self):
        self.taller_abierto = True
        self.autos_en_espera = 5  # Empieza con algo de trabajo
        print(f"\n[SISTEMA]: ¡El mecánico {self.nombre} ha abierto el taller!")
        print(f"Especialidad del día: Reparación de {self.especialidad}")
        print(f"Preparando la {self.herramienta_favorita} para el primer cliente...")

    def mostrar_estado_taller(self):
        estado = "ABIERTO (Trabajando)" if self.taller_abierto else "CERRADO (Descansando)"
        print(f"\n--- FICHA DEL TALLER ---")
        print(f"Mecánico a cargo: {self.nombre}")
        print(f"Especialidad: {self.especialidad}")
        print(f"Herramienta clave: {self.herramienta_favorita}")
        print(f"Estado actual: {estado}")
        if self.taller_abierto:
            print(f"Autos en fila: {self.autos_en_espera}")
        print("-" * 25)

print("¡Bienvenido al Simulador de Taller Mecánico!")
print("Por favor, configura tu perfil profesional:")

# Entradas del usuario
nombre_usuario = input("¿Cuál es tu nombre de mecánico?: ")
rama = input("¿Cuál es tu especialidad? (Motores/Frenos/Electrónica): ")
equipo = input("¿Cuál es tu herramienta favorita? (Llave inglesa/Escáner/Gato hidráulico): ")

# Instancia de la clase
mi_mecanico = Mecanico(nombre_usuario, rama, equipo)

mi_mecanico.mostrar_estado_taller()

opcion = input(f"\n¿Deseas levantar la cortina y empezar a trabajar, {nombre_usuario}? (si/no): ").lower()

if opcion == "si":
    mi_mecanico.abrir_taller()
    mi_mecanico.mostrar_estado_taller()
else:
    print(f"\nEntendido. {nombre_usuario} se toma el día libre. ¡A descansar!")

print("\n" + "="*35)
print("Simulación terminada")
print("Programa terminado")
print("Programa realizado por Meredith Aguirre NL= 2")
```
